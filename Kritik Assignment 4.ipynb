```python
import matplotlib.pyplot as plt  # Import necessary libraries
import numpy as np

def estimate_derivative(func, point):  # Approximate derivative using symmetric difference quotient
    delta = 10**(-10)
    return (func(point + delta) - func(point - delta)) / (2 * delta)

def optimize_function(func, step_size, start_point):
    x_vals = []  # Store x-values
    y_vals = []  # Store corresponding y-values
    x_vals.append(start_point)
    y_vals.append(func(start_point))
    curr_x = start_point
    iteration = 1
    
    while abs(estimate_derivative(func, curr_x)) > 0.001:
        curr_x = x_vals[iteration - 1] - step_size * estimate_derivative(func, x_vals[iteration - 1])
        curr_y = func(curr_x)
        x_vals.append(curr_x)
        y_vals.append(curr_y)
        iteration += 1
        
        if iteration == 10000:  # Prevent infinite loops
            break
    
    plot_range = np.linspace(min(x_vals) - 0.5, max(x_vals) + 0.5, 10000)
    function_values = [func(i) for i in plot_range]
    plt.plot(plot_range, function_values)  # Plot the function
    plt.plot(x_vals, y_vals)  # Plot gradient descent steps
    
    return round(x_vals[-1], 3), round(y_vals[-1], 3)  # Return final values rounded to 3 decimal places

# Example function:
def g(x):
    return abs(x)

print(optimize_function(g, 0.2, 1))

# The function doesn't converge properly at x = 0 due to the non-differentiability of abs(x),
# causing the algorithm to potentially run indefinitely without the iteration limit.
```
